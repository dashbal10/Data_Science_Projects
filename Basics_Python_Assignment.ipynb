```
# Exercise 1: Prime Numbers
# A prime number is a number greater than 1 and divisible only by 1 and itself.

def is_prime(n):
    """Check if the given number is prime."""
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):  # Efficient check up to square root of n
        if n % i == 0:
            return False
    return True

# Test the function
number = int(input("Enter a number to check if it's prime: "))
if is_prime(number):
    print(f"{number} is a Prime Number.")
else:
    print(f"{number} is NOT a Prime Number.")
```

```
# Exercise 2: Product of Random Numbers
import random

# Generate two random integers between 1 and 10
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Ask the user to calculate the product
user_answer = int(input(f"What is the product of {num1} and {num2}? "))

# Check the answer
correct_answer = num1 * num2
if user_answer == correct_answer:
    print("Correct! Well done.")
else:
    print(f"Incorrect. The correct answer is {correct_answer}.")
```

```
# Exercise 3: Squares of Even/Odd Numbers
# Choice: EVEN numbers from 100 to 200

print("Squares of EVEN numbers from 100 to 200:")
for number in range(100, 201):
    if number % 2 == 0:
        print(f"{number}^2 = {number**2}")
```

```
# Exercise 4: Word Counter
# Count the frequency of each word in the input text

input_text = "This is a sample text. This text will be used to demonstrate the word counter."

# Remove punctuation and convert to lowercase
import string

# Clean text
translator = str.maketrans('', '', string.punctuation)
clean_text = input_text.translate(translator).lower()

# Split into words
words = clean_text.split()

# Count occurrences
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Display the result
print("Word Frequency:")
for word, count in word_count.items():
    print(f"'{word}': {count}")
```

```
# Exercise 5: Check for Palindrome

def is_palindrome(s):
    """Return True if string is a palindrome; False otherwise."""
    import string
    # Normalize: remove punctuation, spaces, and convert to lowercase
    translator = str.maketrans('', '', string.punctuation + " ")
    cleaned = s.translate(translator).lower()
    return cleaned == cleaned[::-1]

# Test the function
test_string = input("Enter a string to check for palindrome: ")
if is_palindrome(test_string):
    print("The string is a palindrome.")
else:
    print("The string is NOT a palindrome.")
```
